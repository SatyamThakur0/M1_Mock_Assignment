```python
import pandas as pd

# 1. Load dataset

df = pd.DataFrame({
    "name": ["Amit", "Neha", "Ravi", "Pooja", "Karan", "Anjali", "Rahul", "Sneha"],
    "subject": ["Math", "Math", "Science", "Science", "English", "English", "Math", "Science"],
    "score": [78, 92, 85, 95, 70, 88, 60, 90],
    "study_hours": [5, 7, 6, 8, 4, 6, 3, 7],
    "gender": ["M", "F", "M", "F", "M", "F", "M", "F"]
})

print("Dataset Preview:")
print(df.head())


# 2A. Filtering
# students scoring above 85
high_scorers = df[df["score"] > 85]

print("\nStudents scoring above 85:")
print(high_scorers)


# 2B. Grouping
# average score by subject
avg_score_subject = df.groupby("subject")["score"].mean()

print("\nAverage score by subject:")
print(avg_score_subject)


# 2C. Sorting
# sort students by highest score
sorted_scores = df.sort_values(by="score", ascending=False)

print("\nStudents sorted by score:")
print(sorted_scores)


# 3. Insights
print("\nINSIGHTS:")

# Insight 1
top_subject = avg_score_subject.idxmax()
print(f"1. Subject with highest average score: {top_subject}")

# Insight 2
top_student = sorted_scores.iloc[0]["name"]
print(f"2. Top performing student overall: {top_student}")
```

```python
import matplotlib.pyplot as plt




avg_scores = df.groupby("subject")["score"].mean()

plt.figure()
avg_scores.plot(kind="bar")

plt.title("Average Score by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")

plt.show()



sorted_df = df.sort_values("study_hours")

plt.figure()
plt.plot(sorted_df["study_hours"], sorted_df["score"])

plt.title("Score vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Score")

plt.show()
```
